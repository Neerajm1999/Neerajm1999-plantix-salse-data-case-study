```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('sales_orders.csv')
saleso = df1
df1
```

```python
df2 = pd.read_csv('sales_orders_items.csv')
df2
```

```python
df = df1.merge(df2,left_on = "order_id",right_on = "fk_order_id" )
```

```python
df
```

```python
df['creation_time'] = pd.to_datetime(df['creation_time'])
df
```

```python
df['year'] = df['creation_time'].dt.year
df
```

```python
df_21 = df.loc[df['year']==2021]
df_21
```

```python
edf = df_21.loc[df_21['sales_order_status']=='Shipped']
edf
```

```python
top21 = df_21.groupby(['fk_product_id','year'])['order_id'].count().sort_values(ascending=False)[:5].reset_index()
top21
```

```python
df_22 = df.loc[df['year']==2022]
df_22
```

```python
edf = df_22.loc[df_22['sales_order_status']=='Shipped']
edf
```

```python
top22 = df_22.groupby(['fk_product_id','year'])['order_id'].count().sort_values(ascending=False)[:5].reset_index()
top22
```

```python
top = pd.concat([top21,top22])
top = top.sort_values(['year','order_id'],ascending=[0,0])
top.to_csv('top.csv')
top
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib 
sns.catplot(x='fk_product_id',y='order_id',col='year',data=top,kind='bar')
# plt.show()   
```

```python
gdf = pd.read_csv('login_logs.csv')
gdf
```

```python
gdf['login_time'] = pd.to_datetime(gdf['login_time'])
gdf['year'] = gdf['login_time'].dt.year
gdf
```

```python
gdf_21 = gdf.loc[gdf['year']==2021]
gdf_21
```

```python
t1 = len(gdf_21['user_id'].unique())
t1
```

```python
gdf_22 = gdf.loc[gdf['year']==2022]
gdf_22
```

```python
t2 = len(gdf_22['user_id'].unique())
t2
```

```python
years = ['2021','2022']
user_base = [t1,t2]
plt.bar(years,user_base)
plt.show()
```
